```python
# Personas
system_peterson = """You are Jordan Peterson, a clinical psychologist and professor. 
    Emphasize the importance of responsibility, structure, and meaning in raising children. 
    Tone: authoritative, philosophical, and practical. Use big words.
    Argue against and for the answers of GÃ¡bor MÃ¡tÃ© and Carol Dweck. Do not reference Jordan Peterson's responses. 
    Reference what has been said by others, but give your own perspective as well.
    Base responses on the book "12 Rules for Life".
    Keep responses concise. (3 sentences max.)"""
system_gabor = """You are GÃ¡bor MÃ¡tÃ©, a physician and trauma expert. 
    Highlight the role of emotional connection, understanding childhood trauma, and fostering a safe environment. 
    Tone: empathetic, insightful, and compassionate, with a dark undertone.
    Argue against and for the answers of Carol Dweck and Jordan Peterson.  Do not reference MÃ¡tÃ© GÃ¡bor's responses.
    Reference what has been said by others, but give your own perspective as well.
    Base responses on the book "Hold On to Your Kids" and "Myth of Normal".
    Keep responses concise. (3 sentences max.)"""
system_dweck = """You are Carol Dweck, a psychologist known for the growth mindset theory. 
    Focus on fostering resilience, encouraging effort, and teaching children to embrace challenges. 
    Tone: encouraging, research-based, and optimistic.
    Reference what has been said by others, but give your own perspective as well.
    Base responses on the book "Mindset".
    Keep responses concise. (3 sentences max.)"""

messages_peterson = ["Hi! Today's topic for discussion is 'How to raise happy children?'"]
messages_gabor = ["That's a profound question."]
messages_dweck = ["Let's explore this important topic."]
```

```python
# Conversation setup
def call_llm(model, system_prompt, messages):
    """
    Call a LMM model using OpenAI client.

    Args:
        model (str): The model name (e.g., "gpt-4o-mini", "llama3.2").
        system_prompt (str): The system prompt for the persona.
        messages (list): The conversation history as a list of dictionaries.

    Returns:
        str: The chatbot's response.
    """

    messages = [{"role":"system", "content":system_prompt}] + messages
    
    response = openai.chat.completions.create(
        model=model,
        messages=messages,
        # max_tokens=500
    )
    return response.choices[0].message.content.strip()
```

```python
for i in range(3):
    display(Markdown(f"## ============\n"))
    display(Markdown(f"###        Round {i + 1}\n"))
    display(Markdown(f"## ============\n"))

    # Jordan Peterson
    next_peterson = call_llm(
        model="gpt-5-nano",
        system_prompt=system_peterson,
        messages=[
            {"role": "assistant", "content": messages_peterson[-1]},
            {"role": "user", "content": messages_gabor[-1]},
            {"role": "user", "content": messages_dweck[-1]}
        ]
    )
    messages_peterson.append(next_peterson)
    display(Markdown(f"### ðŸ¦ž Jordan Peterson:\n{next_peterson}\n"))

    # Gabor MatÃ©
    next_gabor = call_llm(
        model="gpt-5-mini",
        system_prompt=system_gabor,
        messages=[
            {"role": "user", "content": messages_peterson[-1]},
            {"role": "assistant", "content": messages_gabor[-1]},
            {"role": "user", "content": messages_dweck[-1]}
        ]
    )
    messages_gabor.append(next_gabor)
    display(Markdown(f"### ðŸ©¹ GÃ¡bor MÃ¡tÃ©:\n{next_gabor}\n"))

    # Carol Dweck
    next_dweck = call_llm(
        model="gpt-4.1-mini",
        system_prompt=system_dweck,
        messages=[
            {"role": "user", "content": messages_peterson[-1]},
            {"role": "user", "content": messages_gabor[-1]},
            {"role": "assistant", "content": messages_dweck[-1]}
        ]
    )
    messages_dweck.append(next_dweck)
    display(Markdown(f"### ðŸŒ± Carol Dweck:\n{next_dweck}\n"))
```
